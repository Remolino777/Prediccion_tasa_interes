```python
import pandas as pd
import lightgbm as lgb
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

### ML Bogota

```python
data = pd.read_parquet(r'D:\0_Respaldo\0_Proyectos_2024\ML_proyects\Prediccion tasa de interes\Prediccion_tasa_interes\data\Transform\train\train_Bogotá.parquet')
data.head(3)
```

```python
df =  data.copy()
```

```python
# Preparacion de los datos
x = df.drop(['Target'], axis=1) #Features
y = df['Target'] #target

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 28)
```

```python
model = lgb.LGBMRegressor()

model.fit(
    x_train, y_train,
    eval_set=[(x_test, y_test)],
    eval_metric='rmse',    
)
```

```python
#Model prediction

y_pred = model.predict(x_test)
predictions = model.predict(x_test)

rmse = mean_squared_error(y_test, y_pred, squared=False)
test_mae = mean_absolute_error(y_test, predictions)

print(f'{test_mae=:.4f}')
print(f'RMSE: {rmse=:.4f}')
```

```python
import joblib as jb
```

```python
jb.dump(model, r'D:\0_Respaldo\0_Proyectos_2024\ML_proyects\Prediccion tasa de interes\Prediccion_tasa_interes\models\ml_bogota.pkl' )
```
